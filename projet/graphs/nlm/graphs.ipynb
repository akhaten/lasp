```python
import matplotlib.pyplot
import matplotlib.cm

import matplotlib

matplotlib.rcParams['figure.facecolor'] = '303030'
matplotlib.rcParams['axes.facecolor'] = '303030'
matplotlib.rcParams['text.color'] = 'FFFFFF'
matplotlib.rcParams['xtick.color'] = 'FFFFFF'
matplotlib.rcParams['ytick.color'] = 'FFFFFF'
matplotlib.rcParams['axes.labelcolor'] = 'FFFFFF'


import numpy
import pandas
```

```python
NAME = 'metrics_1_20_10'

INF = 1
SUP = 20
NB_DENOISE = 10

sigma = numpy.arange(INF, SUP, (SUP-INF)/NB_DENOISE)

df = pandas.read_pickle(NAME+'.pkl')


figure = matplotlib.pyplot.figure(figsize=(10, 10), dpi=100)
# figure.set_facecolor('303030')
fontsize = 30
_ = matplotlib.pyplot.title('MSE', fontsize=fontsize)
_ = matplotlib.pyplot.plot(sigma, df['mse'])
_ = matplotlib.pyplot.xlabel(xlabel='$\sigma$', fontsize=fontsize)
_ = matplotlib.pyplot.ylabel(ylabel='mse', fontsize=fontsize)
index_min = numpy.argmin(df['mse'])
i, j = sigma[index_min], df['mse'][index_min]
_ = matplotlib.pyplot.plot(i, j, 'ro')
matplotlib.pyplot.annotate('({:.3f}, {:.3f})'.format(i, j), xy=(i,j), xytext=(30,0), textcoords='offset points')
_ = matplotlib.pyplot.savefig(NAME+'_mse'+'.jpg')

figure = matplotlib.pyplot.figure(figsize=(10, 10), dpi=100)
# _ = matplotlib.pyplot.subplot(1, 3, 1)
_ = matplotlib.pyplot.title('MAE', fontsize=fontsize)
_ = matplotlib.pyplot.plot(sigma, df['mae'])
_ = matplotlib.pyplot.xlabel(xlabel='$\sigma$', fontsize=fontsize)
_ = matplotlib.pyplot.ylabel(ylabel='mae', fontsize=fontsize)
index_min = numpy.argmin(df['mae'])
i, j = sigma[index_min], df['mae'][index_min]
_ = matplotlib.pyplot.plot(i, j, 'ro')
matplotlib.pyplot.annotate('({:.3f}, {:.3f})'.format(i, j), xy=(i,j), xytext=(30,0), textcoords='offset points')
_ = matplotlib.pyplot.savefig(NAME+'_mae'+'.jpg')


figure = matplotlib.pyplot.figure(figsize=(10, 10), dpi=50)
_ = matplotlib.pyplot.title('PSNR', fontsize=fontsize)
_ = matplotlib.pyplot.plot(sigma, df['psnr'])
_ = matplotlib.pyplot.xlabel(xlabel='$\sigma$', fontsize=fontsize)
_ = matplotlib.pyplot.ylabel(ylabel='psnr', fontsize=fontsize)
index_min = numpy.argmax(df['psnr'])
i, j = sigma[index_min], df['psnr'][index_min]
_ = matplotlib.pyplot.plot(i, j, 'ro')
matplotlib.pyplot.annotate('({:.3f}, {:.3f})'.format(i, j), xy=(i,j), xytext=(30,0), textcoords='offset points')
_ = matplotlib.pyplot.savefig(NAME+'_psnr'+'.jpg')
```
