```python
import matplotlib.pyplot
import matplotlib.cm
import matplotlib

# matplotlib.rcParams['figure.background'] = '303030'
matplotlib.rcParams['figure.facecolor'] = '303030'
matplotlib.rcParams['axes.facecolor'] = '303030'
matplotlib.rcParams['text.color'] = 'FFFFFF'
matplotlib.rcParams['xtick.color'] = 'FFFFFF'
matplotlib.rcParams['ytick.color'] = 'FFFFFF'
matplotlib.rcParams['axes.labelcolor'] = 'FFFFFF'

import numpy
import pandas

import pathlib
```

```python
import lasp.convert
import lasp.metrics
import lasp.noise
import lasp.io
import lasp.filters
```

```python
path_results =  pathlib.Path('./results')
path_output = path_results / pathlib.Path('nlm')
if not(path_output.exists()):
    path_output.mkdir()
```

```python
# Read greylevel image
img = lasp.io.read(image_path = path_results / pathlib.Path('original.npy'))
img_noised = lasp.io.read(image_path = path_results / pathlib.Path('noised.npy'))
```

```python
metrics = {
    'mae' : [],
    'mse' : [],
    'psnr' : []
}

INF = 1
SUP = 20
NB_DENOISE = 10

sigma = numpy.arange(INF, SUP, (SUP-INF)/NB_DENOISE)
```

```python
for sig in sigma:
    img_denoised = lasp.filters.non_local_mean(img_noised, sig, size=5)
    metrics['mae'].append(lasp.metrics.mean_absolute_error(img_denoised, img))
    metrics['mse'].append(lasp.metrics.mean_squared_error(img_denoised, img))
    metrics['psnr'].append(lasp.metrics.peak_signal_to_noise_ratio(img_denoised, img))
    # matplotlib.pyplot.imsave('nlm_{}.png'.format(sig), img_denoised)

df = pandas.DataFrame(metrics)
df.to_pickle(path_output / 'metrics_{}_{}_{}.pkl'.format(INF, SUP, NB_DENOISE))
```
